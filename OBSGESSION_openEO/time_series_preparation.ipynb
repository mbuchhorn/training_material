#### import the needed Python packages

```python
import os
import xarray
import rioxarray
import matplotlib.pyplot as plt
import pandas as pd

from eo_processing.utils.helper import init_connection
from eo_processing.openeo.preprocessing import extract_S1_datacube
from eo_processing.openeo.processing import generate_master_feature_cube, generate_S2_indices
from eo_processing.config import get_job_options, get_collection_options, get_advanced_options, get_standard_processing_options
from eo_processing.utils.geoprocessing import bbox_area, geoJson_2_BBOX
from eo_processing.utils.jobmanager import get_AOI_interactive
```

#### connect to the openEO processing backend

```python
backend = 'cdse' # also possible: terrascope, development, creodias, sentinelhub
# establish the connection to the selected backend
connection = init_connection(backend)
# specify your personal output folder
out_root = r'C:\Users\BUCHHORM\Downloads'
```

#### specify the AOI
Please note: for test reasons we limit the AOIs to 400km2

```python
# we specify the AOI interactive (you can also just load a GeoJson, shapely polygon or set a openEO BBOX manually)
get_AOI_interactive(map_center=(52.2, 6.9), zoom=11)
```

```python
# we convert the drawn or loaded Polygon(s) into a opneEO BoundingBox object
AOI_bbox = geoJson_2_BBOX(os.path.join(out_root, 'polygon.geojson'), delete_file=True, size_check=400)
print(f"The bounding box is: {AOI_bbox} with an area of {bbox_area(AOI_bbox, only_number=True)} km2")
```

## First use case: Sentinel-1 sigma-naught monthly composites 2024
All we have to do is to specify the `processing` and `collection` options for the `eo_processing` package and the `job` options for the openEO backend. Moreover, we have to set the time domain. And finally, use the correct datacube generation function (pipeline) to run in openEO.

```python
# define the options
processing_options = get_advanced_options(provider=backend,
                                          target_crs=3035,
                                          resolution=10.,
                                          ts_interpolation=False,
                                          ts_interval='month')
collection_options = get_collection_options(provider=backend)
job_options = get_job_options(provider=backend, task='raw_extraction')
# the time context is given by start and end date
start = '2024-01-01'
end = '2025-01-01'   # the end is always exclusive
```

```python
# define the openEO pipeline to use
data_cube = extract_S1_datacube(connection,
                                AOI_bbox,
                                start,
                                end,
                                **collection_options,
                                **processing_options)
# create openEO job and run
job = data_cube.execute_batch(title=f'S1 datacube for AOI',
                              out_format='NetCDF',
                              job_options=job_options)
# Get the results and save to output file
results = job.get_results()
outputfile = os.path.join(out_root, f'S1_timeseries_2024.nc')
results.download_file(outputfile)
```

##### visualization of results

```python
ts1 = xarray.load_dataset(outputfile)
ts1
```

```python
# Visualize the time points
ts1.VV.plot(x="x", y="y", col="t", col_wrap=4)
```

## Second use case: Sentinel-2 NDVI seasonalcomposites 2020  2024
All we have to do is to adapt the `processing` and `collection` options for the `eo_processing` package and the `job` options for the openEO backend. And finally, use the correct datacube generation function (pipeline) to run in openEO.

```python
processing_options = get_advanced_options(provider=backend,
                                          target_crs=3035,
                                          resolution=10.,
                                          ts_interpolation=False,
                                          ts_interval='season',
                                          S2_bands=['B04', 'B08'], 
                                          optical_vi_list=['NDVI'], 
                                          append=False)
collection_options = get_collection_options(provider=backend)
job_options = get_job_options(provider=backend, task='vi_generation')
# the time context is given by start and end date
start = '2020-01-01'
end = '2025-01-01'   # the end is always exclusive
```

```python
# define the openEO pipeline to use
data_cube = generate_S2_indices(connection, AOI_bbox, start, end, **collection_options, **processing_options)
# create openEO job and run
job = data_cube.execute_batch(title=f'S2 NDVI datacube for AOI',
                              out_format='NetCDF',
                              job_options=job_options)
# Get the results and save to output file
results = job.get_results()
outputfile = os.path.join(out_root, f'S2_NDVI_timeseries_2020-2024.nc')
results.download_file(outputfile)
```

##### visualization of results

```python
ts2 = xarray.load_dataset(outputfile)
ts2
```

```python
# Visualize 
fig, (axNDVI, axhist) = plt.subplots(1, 2, figsize=(14,7))
ts2.NDVI.sel(t = "2024-06-01").plot(ax=axNDVI)
ts2.NDVI[:,150,150].plot(ax=axhist, ylim=(-1,1), marker='x', color='r', linestyle='--')
plt.show()
```

## Third use case: annual features (metrics) generation for ML tasks like habitat mapping
All we have to do is to adapt the `processing` and `collection` options for the `eo_processing` package and the `job` options for the openEO backend. And finally, use the correct datacube generation function (pipeline) to run in openEO.

```python
processing_options = get_standard_processing_options(provider=backend, task='feature_generation') #standard create over 230 features
collection_options = get_collection_options(provider=backend)
job_options = get_job_options(provider=backend, task='feature_generation')
# the time context is given by start and end date
start = '2024-01-01'
end = '2025-01-01'   # the end is always exclusive
```

```python
# define the openEO pipeline to use
data_cube = generate_master_feature_cube(connection, AOI_bbox, start, end, **collection_options, **processing_options)
# create openEO job and run
job = data_cube.execute_batch(title=f'ML feature datacube for AOI',
                              out_format='GTiff',
                              job_options=job_options)
# Get the results and save to output file
results = job.get_results()
outputfile = os.path.join(out_root, f'feature_datacube_2024.tif')
results.download_file(outputfile)
```

##### visualization of results

```python
data = rioxarray.open_rasterio(outputfile, band_as_variable=True, masked=True)
data = data.rename({band:data[band].attrs["long_name"] for band in data})
data
```

```python
# visualize just the BLFEI features as faceted plot
arrlist = []
lMedian = [x for x in data.keys() if ('BLFEI' in x) and ('_sum' not in x)]
for varname in lMedian:
    tmp      = data[varname]
    arrlist.append(tmp)
xrda4D = xarray.concat(arrlist, pd.Index(lMedian, name='feat'))
xrda4D.plot(col="feat", col_wrap=4, )
```
